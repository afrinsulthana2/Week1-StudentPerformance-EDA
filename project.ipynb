# Loading the dataset

```python
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

# Checking basic info about the dataset

```python
df.info()
```

## Getting summary statistics for numerical columns

```python
df.describe()
```

## Check for Missing Values

```python
df.isnull().sum()
```

# Analyze by Gender

```python
df.groupby("gender")[["math score", "reading score", "writing score"]].mean()
```

# Analyze by Test Preparation

```python
df.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean()
```

# Analyze by Lunch Type

```python
df.groupby("lunch")[["math score", "reading score", "writing score"]].mean()
```

# Analyze by Parental Education

```python
df.groupby("parental level of education")[["math score", "reading score", "writing score"]].mean()
```

# Bar plot: Math score by gender

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x="gender", y="math score", data=df, palette="Set3")
plt.title("Average Math Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()
```

# Box plot: Reading score by lunch type

```python
sns.boxplot(x="lunch", y="reading score", data=df, palette="Set2")
plt.title("Reading Score Distribution by Lunch Type")
plt.xlabel("Lunch Type")
plt.ylabel("Reading Score")
plt.show()
```

# Histogram: Writing score distribution

```python
sns.histplot(df["writing score"], bins=20, kde=True, color="purple")
plt.title("Distribution of Writing Scores")
plt.xlabel("Writing Score")
plt.ylabel("Number of Students")
plt.show()
```

### Count plot: Number of students by test preparation course

```python
sns.countplot(x="test preparation course", data=df, palette="dark")
plt.title("Student Count by Test Preparation Course")
plt.xlabel("Course Status")
plt.ylabel("Count")
plt.show()
```

key Insights from Student Performance Analysis:

1. **Test Preparation Helps**  
   Students who **completed the test preparation course** scored significantly **higher** in all three subjects  especially in **reading and writing**.

2. **Female Students Excel in Language**  
   While **male students scored higher in math**, **female students outperformed in reading and writing**.

3. **Lunch Type Influences Performance**  
   Students who had **standard lunch** scored better on average than those who had **free/reduced lunch**, indicating a possible link between nutrition and performance.

4. **Parental Education Makes a Difference**  
   Students whose parents had a **masters or bachelors degree** generally performed better compared to those whose parents had **high school or associate degrees**.

5. **Math Scores Are the Most Variable**  
   The **distribution of math scores** shows greater variation compared to reading and writing, based on histogram and box plot analysis.
